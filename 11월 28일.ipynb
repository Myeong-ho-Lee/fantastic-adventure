```python
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010] 
gdp = [67.0, 80.0, 257.0, 1686.0, 6505, 11865.3, 22105.3]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(years, gdp, color = 'b', marker ='o', linestyle = 'dashed', label = 'gdp of korea')

plt.title('GDP per capita')
plt.ylabel('dollors')
plt.xlabel('years')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


x = np.linspace(-20, 20, 41)
y1 = 2 * x
y2 = x**2 + 5
y3 = - x ** 2 - 5

plt.plot(x, y1, 'r--', label = 'y = 2x')
plt.plot(x, y2, 'g^-', label = 'y= x^2+5')
plt.plot(x, y3, 'b*:', label = 'y = -x^2-5')

plt.legend()
plt.grid()
plt.xlim([-30, 30])
plt.ylim([-30, 30])
plt.show()
```

```python
theta = np.linspace(0, 360, 361)
sin = np.sin(2*np.pi*theta/360)
cos = np.cos(2*np.pi*theta/360)

plt.plot(theta, sin, "k-", label = 'sin')
plt.plot(theta, cos, "r--", label = 'cos')
plt.legend()
plt.xlabel('theta')
plt.ylabel('values')
plt.grid()
plt.title('SIN/COS wave')
plt.show()
```
